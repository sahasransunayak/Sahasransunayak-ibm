```python
import numpy as np
import matplotlib.pyplot as plt

# Generate binomial data
x = np.random.binomial(n=10, p=0.5, size=10)

# Create a histogram
plt.hist(x, bins=range(12), edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Create a dictionary with student details
students = {
    "name": [
        "Alice Johnson", "Bob Smith", "Charlie Brown", "Diana Prince",
        "Ethan Hunt", "Fiona Gallagher", "George Smith", "Hannah Baker"
    ],
    "id": [
        "S001", "S002", "S003", "S004",
        "S005", "S006", "S007", "S008"
    ],
    "department": [
        "Computer Science", "Mathematics", "Physics", "Chemistry",
        "Engineering", "Biology", "History", "Literature"
    ],
    "grade": [
        "A", "B", "A-", "B+",
        "B-", "A", "C+", "A+"
    ]
}

# Convert dictionary into a DataFrame
df = pd.DataFrame(students)

# 1. Retrieve students using loc for a specific department
def get_students_by_department(department):
    return df.loc[df['department'] == department]

# Example usage
cs_students = get_students_by_department("Computer Science")
print("Computer Science Students:\n", cs_students)

# 2. Retrieve rows using iloc for rows 2 and 3
rows_2_3 = df.iloc[2:4]  # iloc retrieves rows by integer index
print("\nRows 2 and 3:\n", rows_2_3)

# 3. Retrieve the IDs of the students using loc
student_ids = df.loc[:, 'id'].tolist()  # Using loc to get the 'id' column
print("\nStudent IDs:\n", student_ids)
```
